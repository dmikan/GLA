```python
import numpy as np
import matplotlib.pyplot as plt

# Define q_gl_max
q_gl_max = 4000

# Generate data
data = np.logspace(0.1, np.log10(q_gl_max), 1000)

# Create the plot
plt.figure(figsize=(12, 3))

# Plot the points on a single line (y=1)
plt.plot(data, np.ones_like(data), 'o', markersize=3, alpha=0.5)

# Set title and labels
plt.title(f'Espaciamiento de Valores con $\\text{{np.logspace}}(0.1, \\log_{{10}}({q_gl_max}), 1000)$', fontsize=10)
plt.xlabel('Valor (Escala Lineal)', fontsize=10)

# Hide the y-axis as it only represents a single line of points
plt.yticks([])
plt.ylim(0.9, 1.1)

# Ensure the plot is tight and looks good
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.savefig('logspace_spacing_linear.png')
print("Se ha generado el gráfico 'logspace_spacing_linear.png'")

# Also show on a log scale for better visualization of the log-spacing
plt.figure(figsize=(12, 3))
plt.plot(data, np.ones_like(data), 'o', markersize=3, alpha=0.5)
plt.xscale('log')

# Set title and labels
plt.title(f'Espaciamiento de Valores con $\\text{{np.logspace}}(0.1, \\log_{{10}}({q_gl_max}), 1000)$', fontsize=10)
plt.xlabel('Valor (Escala Logarítmica)', fontsize=10)

# Hide the y-axis as it only represents a single line of points
plt.yticks([])
plt.ylim(0.9, 1.1)

# Ensure the plot is tight and looks good
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.savefig('logspace_spacing_log.png')
print("Se ha generado el gráfico 'logspace_spacing_log.png'")
```
